```python
from datetime import datetime,timedelta,date
from dateutil.relativedelta import relativedelta, MO
import re,sys
```

```python
def getWeekId():
    '''getting weekId'''
    today = date.today()
    last_monday = today + relativedelta(weekday=MO(-1))
    dt = last_monday.strftime("%Y/%m/%d")
    regex = '^\d{4}/\d{2}/\d{2}$'

    for argv in sys.argv:
        if len(re.findall(regex, argv)) != 0:
            dt = argv

    dt = datetime.strptime(dt, "%Y/%m/%d") - timedelta(weeks=1) 

    weekId_list = []
    for i in range(144):
        dt = dt - timedelta(days = 7)    
        if len(str(dt.isocalendar().week)) == 2 :
            weekId = "{}{}".format(dt.isocalendar()[0], dt.isocalendar()[1]) 
            weekId_list.append(weekId)
        else:
            weekId = "{}0{}".format(dt.isocalendar()[0], dt.isocalendar()[1]) 
            weekId_list.append(weekId)


    return weekId_list
```

```python
for weekid in getWeekId():
    print(weekid)
```
